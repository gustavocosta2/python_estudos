# **Gráfico de Dispersão (Scatter plots)**

O gráfico de dispersão mostra a relação entre duas variáveis quantitativas. Cada ponto no gráfico representa um par de valores (x,y).

**USO:** Recomendado para identificar tendências, correlações ou padrões nos dados. Por exemplo, pode mostrar a relação entre o tempo de estudo e as notas dos alunos.

## Elementos do Gráfico de Dispersão

- **Eixos:** O eixo horizontal (x) e o eixo vertical (y) representam as duas variáveis que estão sendo comparadas.
- **Pontos:** Cada ponto representa um par de valores das duas variáveis.
- **Padrão de Pontos:** A dispersão dos pontos pode indicar a natureza da relação entre duas variáveis (linear, exponencial, nenhuma relação).

````python
# Importação da biblioteca matplotlib e da biblioteca seaborn para plotagem gráfica.
import matplotlib.pyplot as plt
import seaborn as sns

# Importação da biblioteca científica NumPy para a criação dos dados do gráfico.
import numpy as np

# Dados para a plotagem do Gráfico de Dispersão.
idade = np.array([25, 30, 35, 40, 45, 50, 55, 60, 65])
salario = np.array([40000, 45000, 50000, 55000, 60000, 65000, 70000, 75000, 80000])

# Plotagem gráfica.
sns.scatterplot(x = idade, y = salario, color = 'blue')
plt.title('Relação entre Idade e Salário')
plt.xlabel('Idade')
plt.ylabel('Salário Anual ($)')
````
