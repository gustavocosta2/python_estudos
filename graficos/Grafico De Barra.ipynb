# Gráfico de Barras

O gráfico de barras utiliza barras (também chamadas de bins) para mostrar comparações entre categorias de dados. Cada barra representa um grupo definido em termos de uma variável categórica. Ele é muito utilizado para comparar quantidades em diferentes categorias.

**USO:** Ideal para comparar quantidades em diferentes categorias, como vendas de diferentes produtos.

## Elementos do Gráfico de Barras

- **Altura da Barra:** Representa a medida da variável (como vendas, pontuação, etc.).
- **Largura da Barra:** Geralmente uniforme (mesmo tamanho), representando categorias distintas.
- **Espaçamento entre Barras:** Os espaços entre as barras evidenciam a independência entre as categorias.

````python
# Importação da matplotlib que é uma biblioteca de plotagem gráfica.
import matplotlib.pyplot as plt

# Importação da biblioteca seaborn para plotagem gráfica.
import seaborn as sns

# Dados para o gráfico de barras horizontais.
empresas = ['Empresa A', 'Empresa B', 'Empresa C', 'Empresa D']
lucros = [300, 450, 500, 350]

# Plotagem gráfica do Gráfico de Barras.
sns.barplot(x = lucros, y = empresas, palette = 'coolwarm', hue =  empresas)
plt.title('Lucro por Empresa')
plt.xlabel('Lucro (em milhares de dólares)')
plt.ylabel('Empresa')
````

## Exercícios Práticos

### **1. Comparação de Vendas por Produto**

Você tem um conjunto de dados que mostra as vendas de diferentes produtos em um mês. Crie um gráfico de barras que mostre as vendas de cada produto.

```
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
vendas = [150, 90, 200, 130, 170]
```

Tarefas:

- Crie um gráfico de barras simples que mostre as vendas de cada produto.
- Adicione rótulos aos eixos X e Y.
- Adicione um título ao gráfico.

````python
# Importação das bibliotecas necessárias.
import matplotlib.pyplot as plt
import seaborn as sns

# Conjunto de dados para a plotagem.
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
vendas = [150, 90, 200, 130, 170]

# Plotagem gráfica.
sns.barplot(
    x = produtos,
    y = vendas,
    hue = produtos, # coloração atrelada à variável 'produtos'.
    palette = 'deep',
    orient = 'v'
)

# Criação dos rótulos dos eixos X e Y.
plt.ylabel('Quantidade vendida')
plt.xlabel('Produto')

# Criação do título do gráfico.
plt.title('Vendas dos Produtos')
````
